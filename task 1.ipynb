```python
#function to drop the duplicate values

def remove_dup(dataframe):
    org_df = dataframe.drop_duplicates()
    return org_df
```

```python
#function to get only the unique values of a column from a certain dataframe

def get_unique_values(dataframe, column):
    col_df = dataframe[[column]]
    for value in col_df:
        unique_values = col_df[value].unique()
    return unique_values    
```

```python
#function to get a dictionary that gives list of unique dobs, count of dobs and total records for given postcode

def get_postcode_data(postcode, dataframe):
    output_kwargs = {
        'post_code': postcode
    }
    pc_data = dataframe.loc[dataframe['post_code'] == postcode] 
    dob_data = get_unique_values(pc_data, 'dob')
    dob_list = list(dob_data)
    output_kwargs['dob'] = dob_list
    output_kwargs['total_unique_dob'] = len(dob_list)
    output_kwargs['total_records'] = pc_data.shape[0]
    return output_kwargs
```

```python
#getting unique values of postcode first
value_list = get_unique_values(df, 'post_code')
value_list
```

```python
#getting the final result for each postcode(task 1)

def get_pc_list():
    op_pc_list = [get_postcode_data(each, df) for each in list(value_list)]
    return op_pc_list
```
