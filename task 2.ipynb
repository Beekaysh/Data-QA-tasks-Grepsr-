```python
import pandas as pd
import numpy as np
```

```python
#Reading the csv file and handling missing values
missing_values = ['na','NA', 'N/A', 'N/a', 'not available', 'missing', np.nan]
df = pd.read_csv('dataset.csv', na_values = missing_values)
```

```python
#fxn to remove duplicates
def remove_dup(dataframe):
    org_df = dataframe.drop_duplicates()
    return org_df

#defining a function to clean the dataset:

def clean_data(dataframe):
    a = remove_dup(dataframe) #using the fxn to remove duplicates
    a = a.dropna(how = 'all') #drop the records that have all na values
    a = a.dropna(axis=1, how='all') #drop the columns that have all na values
    return a
```

```python
#storing all the duplicated/removed values in a separate dataframe

def duplicated_dataset(dataframe):
    dup_df = dataframe.loc[dataframe.duplicated(), :]
    return dup_df
```

```python
dff = clean_data(df)
dff
```

```python
def get_unique_postcode(dataframe, column):
    col_df = dataframe[[column]]
    for value in col_df:
        unique_values = col_df[value].unique()
    return unique_values    
```
